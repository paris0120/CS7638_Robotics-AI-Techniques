### Heuristic Funtion
1. a funtion has to be set up
    1. $h(x,y) \leq$ distance to goal from x,y
1. all zeros then A* resorts back to the search algorithm already implemented
1. not need to be accurate



### How to
1. impliment an open list
1. add our state [f, g, h, x, y]
    1. F value = G value + Heuristic Value
    1. $f = g+h(x,y)$
1. when expand, remove the element with the lowest f-value instead of the g-value

## Dynamic Programming
1. to get optimal policy
1. Given
    1. Map
    1. Goal
1. Output
    1. Best path from anywhere
1. Policy
    1. x, y -> action
1. Value function
    1. $f(x,y) = min f(x',y')+1$
    1. link to the goal (0)
    

## Planning
1. Discrete
    1. A* -> path
    1. DP -> Policy
    

### Admissable Heuristic
1. $h(x) \leq $ cost-to-goal
1. if not

    1. the path may not be optimal
